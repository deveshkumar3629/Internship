```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

```python
cont=page.content
cont
```

```python
soup=BeautifulSoup(page.content)
```

### Scraping first header tag

```python
head=soup.find('span',class_="mw-headline")
```

```python
head
```

```python
head.text
```

### All header

```python
head=[]
for i in soup.find_all('span',class_="mw-headline"):
    head.append(i.text)
    
head
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Header tag':head})
```
